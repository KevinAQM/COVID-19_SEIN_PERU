```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Cargando archivo ---
data = pd.read_csv("data/Demand_Mar-Apr_2020_15min.csv",
                        parse_dates=[0], index_col=0, dayfirst=1)
```

```python
# Observando primeras filas de nuestra data ---
data.head()
```

```python
#Renombrando columnas
data.rename(columns={'Hidro Total':'Hidroeléctrica', 'Termo Total':'Termoeléctrica', 
                     'Eolica Total':'Eólica', 'Solar Total':'Solar', 'Generacion Total':'Generación Total', 
                     'Exportacion':'Exportación', 'Importacion':'Importación' }, inplace = True)
data
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(11,11));

data['Generación Total'].plot(ax=ax[0], title='Generación total\n01 Marzo - 31 Abril 2020', color = 'black')
data['Hidroeléctrica'].plot(ax=ax[1], title='Generación, por tipo\n01 Marzo - 31 Abril 2020', color = 'C0')
data['Termoeléctrica'].plot(ax=ax[1], color = 'darkslategrey')
data['Eólica'].plot(ax=ax[1], color = 'deepskyblue')
data['Solar'].plot(ax=ax[1], color = 'goldenrod')

ax[0].legend(loc='upper right')
ax[1].legend(loc='center right')

# ax[0].text("2020-03-16", 5000, 'Horas Punta', ha='right')

ax[0].annotate("Inicio Cuarentena\nCOVID-19 Perú", xy=('2020-03-16', 4800),  xycoords='data',
            xytext=(-100, -30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"), color='red')
ax[1].annotate("Inicio Cuarentena\nCOVID-19 Perú", xy=('2020-03-16', 3200),  xycoords='data',
            xytext=(-100, -30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"), color='red')

for ax in fig.get_axes():
    ax.xaxis.set_label_text("")
    ax.set_ylabel('Potencia [MW]')
    ax.axvline("2020-03-02", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-03-09", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-03-16", color="red", linestyle="--", lw=1, alpha=1)
    ax.axvline("2020-03-23", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-03-30", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-04-06", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-04-13", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-04-20", color="gray", linestyle="--", lw=1, alpha=0.8)
    ax.axvline("2020-04-27", color="gray", linestyle="--", lw=1, alpha=0.8)
    

plt.tight_layout()
plt.savefig('Generación Total.png', dpi = 200)
```

Vamos a dividir la data en 3 periodos:
    - Antes de la cuarentena (01 Marzo - 16 Marzo 2020)
    - Una semana después del inicio de la cuarentena (16 Marzo - 23 Marzo 2020)
    - Después de la semana de transición (23 Marzo - 31 Abril 2020)


```python
#Separando periodos
before_quarantine = data['2020-03-01' : '2020-03-15']
transition_quarantine = data['2020-03-16' : '2020-03-22']
after_transition_quarantine = data['2020-03-23' : '2020-04-30']
```

```python
#Separando títulos de columnas a analizar en una lista 
columns = ['Generación Total', 'Hidroeléctrica', 'Termoeléctrica', 'Eólica', 'Solar']
```

```python
#Graficando Generación por tipo, incluyendo valores promedio en cada periodo.
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(11,22));

data['Generación Total'].plot(ax=ax[0], title='Generación Total\n01 Marzo - 31 Abril 2020', color = 'black')
data['Hidroeléctrica'].plot(ax=ax[1], title='Generación Hidroeléctrica\n01 Marzo - 31 Abril 2020', color = 'C0')
data['Termoeléctrica'].plot(ax=ax[2], title='Generación Termoeléctrica\n01 Marzo - 31 Abril 2020', color = 'darkslategrey')
data['Eólica'].plot(ax=ax[3], title='Generación Eólica\n01 Marzo - 31 Abril 2020', color = 'deepskyblue')
data['Solar'].plot(ax=ax[4], title='Generación Solar\n01 Marzo - 31 Abril 2020', color = 'goldenrod')

ax[0].annotate("Inicio Cuarentena\nCOVID-19 Perú", xy=('2020-03-16', 4800),  xycoords='data',
            xytext=(-100, -30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"), color='red')

for ax, column in zip(fig.get_axes(), columns):
    ax.xaxis.set_label_text("")
    ax.set_ylabel('Potencia [MW]')
    ax.legend(loc='upper right')
    ax.axvline("2020-03-02", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-03-09", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-03-16", color="red", linestyle="--", lw=1.2, alpha=1)
    ax.axvline("2020-03-23", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-03-30", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-04-06", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-04-13", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-04-20", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axvline("2020-04-27", color="gray", linestyle="--", lw=0.8, alpha=0.6)
    ax.axhline(y=before_quarantine[column].mean(), xmin=0, xmax=0.25, lw=0.8, alpha=0.8, color='black')
    ax.axhline(y=transition_quarantine[column].mean(), xmin=0.25, xmax=0.36, lw=0.8, alpha=0.8, color='black')
    ax.axhline(y=after_transition_quarantine[column].mean(), xmin=0.36, xmax=1, lw=0.8, alpha=0.8, color='black')
    ax.annotate("{:.3f} MW".format(before_quarantine[column].mean(), '2020-03-09'), 
               xy=('2020-03-09', before_quarantine[column].mean()), xycoords='data', 
               xytext=('2020-03-09', before_quarantine[column].mean()), 
               bbox=dict(boxstyle="round", fc="w", ec='0.5', alpha=1), ha='center', va='bottom', fontsize=9)
    ax.annotate("{:.3f}\n(Δ={:.2f}%)".format(transition_quarantine[column].mean(),
                                   (transition_quarantine[column].mean()/before_quarantine[column].mean() - 1)*100), 
               xy=('2020-03-19 12:00', transition_quarantine[column].mean()), xycoords='data', 
               xytext=('2020-03-19 12:00', transition_quarantine[column].mean()), 
               bbox=dict(boxstyle="round", fc="w", ec='0.5', alpha=1), ha='center', va='top', fontsize=9)
    ax.annotate("{:.3f}\n(Δ={:.2f}%)".format(after_transition_quarantine[column].mean(), 
                                   (after_transition_quarantine[column].mean()/before_quarantine[column].mean() - 1)*100), 
               xy=('2020-04-09 12:00', after_transition_quarantine[column].mean()), xycoords='data', 
               xytext=('2020-04-09 12:00', after_transition_quarantine[column].mean()), 
               bbox=dict(boxstyle="round", fc="w", ec='0.5', alpha=1), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.savefig('Generación por Tipo.png', dpi = 250)
```

1. Las variaciones (Δ) son respecto a la media de la data antes del inicio de la cuarentena.
2. Observamos que la media de la Generación Total disminuyó en un tercio (Δ=-33.05%).
3. La generación hidroeléctrica no fue tan afectada (Δ=-14.23%)
4. El gran perdedor fue la generación termoeléctrica, que ahora solo produce como media un 10% de lo que producía antes de la cuarentena. A una menor demanda muchas centrales termoeléctricas (quienes son las que marginan) no despachan.
5. La generación eólica ha subido un 23%.
6. La generación solar se mantiene.

```python
#Nuevo dataframe para stackplot
generation = data[['Solar', 'Eólica', 'Hidroeléctrica', 'Termoeléctrica']]

#Separando Nuevo dataframe por periodos a analizar
gen_before_quarantine = generation['2020-03-01' : '2020-03-15']
gen_transition_quarantine = generation['2020-03-16' : '2020-03-22']
gen_after_transition_quarantine = generation['2020-03-23' : '2020-04-30']

pal=['goldenrod', 'deepskyblue', 'C0', 'darkslategrey']
```

```python
#Stackplot por periodos
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(12,14));

gen_before_quarantine.plot.area(ax=ax[0], stacked=True, color=pal)
ax[0].set_title('Potencia acumulada, antes del inicio de la cuarentena\n01 Marzo - 16 Marzo 2020', fontsize=12)
ax[0].axvline("2020-03-16", color="red", linestyle="--", lw=1.5, alpha=1)
ax[0].annotate("Inicio Cuarentena\nCOVID-19 Perú", xy=('2020-03-16', 3500),  xycoords='data',
            xytext=(-90, -30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2"), color='w', fontsize=9)
ax[0].axvline("2020-03-02", color="gray", linestyle="--", lw=0.8, alpha=0.6)
ax[0].axvline("2020-03-09", color="gray", linestyle="--", lw=0.8, alpha=0.6)

gen_transition_quarantine.plot.area(ax=ax[1], stacked=True, color=pal)
ax[1].set_title('Potencia acumulada, una semana después del inicio de la cuarentena\n16 Marzo - 23 Marzo 2020', fontsize=12)
ax[1].axvline("2020-03-16 00:15", color="red", linestyle="--", lw=1.5, alpha=1)
ax[1].annotate("Inicio Cuarentena\nCOVID-19 Perú", xy=('2020-03-16', 3500),  xycoords='data',
            xytext=(20, -30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2"), color='w', fontsize=9)

gen_after_transition_quarantine.plot.area(ax=ax[2], stacked=True, color=pal)
ax[2].set_title('Potencia acumulada, después de la semana de transición\n23 Marzo - 31 Abril 2020', fontsize=12)
ax[2].axvline("2020-03-30", color="gray", linestyle="--", lw=0.8, alpha=0.6)
ax[2].axvline("2020-04-06", color="gray", linestyle="--", lw=0.8, alpha=0.6)
ax[2].axvline("2020-04-13", color="gray", linestyle="--", lw=0.8, alpha=0.6)
ax[2].axvline("2020-04-20", color="gray", linestyle="--", lw=0.8, alpha=0.6)
ax[2].axvline("2020-04-27", color="gray", linestyle="--", lw=0.8, alpha=0.6)

for ax in fig.get_axes():
    ax.xaxis.set_label_text("")
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax.set_ylabel('Potencia [MW]')
    ax.grid(False)

plt.tight_layout()

plt.savefig('Potencia acumulada por periodos.png', dpi = 250)
```
